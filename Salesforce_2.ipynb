### Working with SQL in Jupyter Notebook

```python
import pandas as pd
import pandasql as ps
```

```python
file_location = "Salesforce 01.xlsx"
car_launches = pd.read_excel(file_location, sheet_name = "car_launches")
```

#### New Car Model Each Year
* Here is a table of the new products of each company in 2019 and 2020. 

```python
car_launches
```

* Output the company name, and the net amount of new products of each company in 2020 over the amount of that in the previous year.

```python
ps.sqldf(
    """
    SELECT company_name,
           COUNT(CASE WHEN year = 2020 THEN product_name END) - COUNT(CASE WHEN year = 2019 THEN product_name END) AS new_products_diff
    FROM car_launches
    GROUP BY company_name
    ORDER BY new_products_diff DESC
    
    """
    )
```
