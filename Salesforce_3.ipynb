#### Find The Staff With The Highest Target Under Manager 1
* Find the staff with the highest target number and also works under manager id 1. 
* Output the first name, surname, and target he has achieved.

```python
ps.sqldf(
    """

    SELECT first_name, last_name, target
    FROM salesforce_employees
    WHERE target = (    
                    SELECT MAX(target) as target
                    FROM salesforce_employees                    
                    )
           AND manager_id = 1

    """
    )
```
