#### Compare each staffâ€™s salary to the average salary of his department.
1. Find out average salary by department. Add it to the table.

```python
ps.sqldf(
    """

    SELECT department, first_name, salary,
           CAST(AVG(salary) OVER(PARTITION BY department) as int) AS avg,
           salary - CAST(AVG(salary) OVER(PARTITION BY department) as int) AS diff
    FROM employee

    """
    )
```
